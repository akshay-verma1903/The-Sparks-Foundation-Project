### Bivariate Analysis

```python
#correlation matrix
data.corr()
```

```python
fig, axes = plt.subplots(1,1, figsize = (8,5))
sns.heatmap(data.corr())
plt.show()
```

#### Here we can see discount and profit are highly negatively corelated

Values closer to zero means there is no linear trend between the two variables.

The close to 1 the correlation is the more positively correlated they are; that is as one increases so does the other and

the closer to 1 the stronger this relationship is. A correlation closer to -1 is similar,

but instead of both increasing one variable will decrease as the other increases

```python
sns.catplot(x='Category',y='Profit',data=data,col="Region")
```

```python
plt.figure(figsize=(10,10))
data.groupby('Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.show()
```

#### From the above graph we can see that the sales of furniture is good still profit is lowest

```python
plt.figure(figsize=(24,10))
data.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.show()
```

```python
plt.figure(figsize=(24,10))
data.groupby('Sub-Category')['Profit'].agg(['sum']).plot.bar()
plt.show()
```

#### From the above figure we can see that:-

1) Sale of tables is high but it is producing loss

2) There is a negligible loss from Bookcases and suppliers

3) There is almost negligible sale and profit from fasteners

4) Copiers are giving the highest profit

```python
sns.barplot(x=data.Discount,y=data.Profit)
```

#### From the above graph we can see that from more than 20% discount loss is happening

```python
sns.barplot(x=data.Region, y=data.Sales)
```

```python
sns.barplot(x=data.Region, y=data.Profit)
```

```python
sns.barplot(x=data.Region, y=data.Profit,hue=data.Category)
```

1) For the central region the Furniture category is making most loss

2) Technology is giving the highest profit in every region

```python
sns.barplot(x=data.Region, y=data.Profit,hue=data.Segment)
```

#### From the above figure we can see that the West region produces highest net profit and East region is producing very high profit in Home office segment while the lowest profit was in Central region in consumer segment

```python
dataplot = data.groupby(['State'])['Sales', 'Profit'].sum()
dataplot.plot.bar(figsize = (20,10))
```

#### From the above we see that there are some states with loss

```python
dataplot = data.groupby(['State'])['Profit'].sum()
dataplot.plot.bar(figsize = (20,10))
```

#### From the above graph we can see that:-

1) The states producing loss are Arizona, Colarado, Florida, illinois, North Carolina, Ohio, Pennsylvania, Tennesee, Texas

2) Texas, Ohio, Illinois and Pennsylvania are incurring huge loss even when the sale is good

3) New York and California are making the highest profit

```python
dataplot = data.groupby(['State']).Discount.sum()
plt.xlabel('States')
plt.ylabel('Discount')
dataplot.plot.bar(figsize = (20,10))
```

#### From the above graph we can see that texas is giving very high discounts and thus incurring huge loss. The same is the case with Illinois, Ohio and Pennsylvania.
